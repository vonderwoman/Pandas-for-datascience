```python
import pandas as pd
import numpy as np
```

```python
Packing_dict={
              0:'Sleepwear',
              1:'Shirts',
              2:'Socks',
              3:'Jeans',
              4:'Legging',
              5:'Garments',
              6:'Skirts',
              7:'Coats',
              8:'Boots',
              9:'Scarves',
              10:'Dress shoes' } 
    

Packing=pd.Series(Packing_dict)
Packing
```

```python
Clothes_dict ={
              0:'Sleepwear',
              1:'Shirts',
              2:'Socks',
              3:'Jeans',
              4:'Legging',
              5:'Garments',
              6:'Skirts',
              7:'Coats',
              8:'Boots',
              9:'Scarves',
              10:'Dress shoes'
    
}
Clothes=pd.Series(Clothes_dict)
Clothes
```

```python
Toiletries_dict={
              0:'Toothbrush',
              1:'Toothpaste',
              2:'Dental floss',
              3:'Soap',
              4:'Shampoo',
              5:'Face cleanser',
              6:'Sunscreen',
              7:'Makeup',
              8:'Brush',
              9:'Comb',
              10:'Lotion'
    
}
Toiletries =pd.Series(Toiletries_dict)
Toiletries
```

```python
Basics_dict={
              0:'Cell phone',
              1:'Charger',
              2:'Passport',
              3:'Laptop',
              4:'Wallet',
              5:'Plug adapter',
              6:'Flight ticket',
              7:'Makeup',
              8:'Insurance',
              9:'Copies of documents',
              10:'Visa'
    
}
Basics =pd.Series(Basics_dict)
Basics
```

```python
Quantity_dict={
              0:45,
              1:30,
              2:40,
              3:50,
              4:2,
              5:30,
              6:56,
              7:34,
              8:22,
              9:23,
              10:45
    
}
Quantity =pd.Series(Quantity_dict)
Quantity
```

```python
InitialPrice_dict={
    0:45,
              1:340,
              2:450,
              3:540,
              4:245,
              5:350,
              6:546,
              7:364,
              8:242,
              9:273,
              10:465
    
}
InitialPrice =pd.Series(InitialPrice_dict)
InitialPrice
```

```python
Packing=pd.DataFrame({
    
    'Clothes':Clothes,
    'Toiletries':Toiletries,
    'Basics':Basics,
    'Quantity':Quantity,
    'InitialPrice':InitialPrice
})
Packing
```

```python
Packing['New Price'] =Packing['Quantity'] /Packing['InitialPrice'] 
```

```python
Packing
```

```python
Packing['percentage'] = (Packing['InitialPrice'] / 
                      Packing['New Price'].sum()) * 100
```
